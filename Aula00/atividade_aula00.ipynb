### Questão 1
**Por que o resultado de math.sqrt(2)**2 não é exatamente 2?**
R: O computador armazena números em formato binário com uma quantidade limitada de bits (precisão finita). Como a raiz de 2 é um número irracional (tem infinitas casas decimais), o computador precisa "arredondar" (truncar) o valor para caber na memória. Ao elevar ao quadrado novamente, esse pequeno erro de arredondamento aparece no final.

### Questão 2
**O resultado de 0.1 + 0.2 invalida o uso do computador na engenharia? Justifique.**
R: Não invalida. Na engenharia, trabalhamos sempre com margens de erro aceitáveis (tolerâncias). O erro apresentado (na ordem de 10^-17) é extremamente pequeno e desprezível para a maioria das aplicações práticas. Além disso, disciplinas como Cálculo Numérico existem justamente para ensinar métodos que controlam e minimizam esses erros.
